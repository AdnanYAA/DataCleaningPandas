# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df=pd.read_excel("Customer Call list.xlsx")
df.head(4)
```

```python
df.isna().sum()
```

```python
#removing the column
df.drop("Not_Useful_Column", inplace=True, axis=1)
df.head(3)
```

```python
#removing duplicates
df=df.drop_duplicates()
df.head()
```

```python
#refining the data of Last-name column

# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")              #We can do this by this way
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")


df["Last_Name"]=df["Last_Name"].str.strip("123._/")
df.head()
```

```python
# preprocessing on Phone_number column

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')

#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df.head()
```

```python
#expand the address into three more columns

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df.head()
```

```python
#did some formatting on Don't_contact column

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df.head(2)
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')


df=df.fillna('')
df.head()
```

```python
#as we don't want those who want that no one contact us. so removing those entries is good option.

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df.head()
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df.head()

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
#more refined dataframe is available now.

df = df.reset_index(drop=True)
df.head()
```
